**FUNGSI REKURSIF**

**Pengertian Fungsi Rekursif**

Fungsi rekursif adalah fungsi yang dapat memanggil dirinya sendiri secara berulang-ulang hingga suatu kondisi yang di definisikan terpenuhi atau bernilai benar(true).

**Aturan dalam fungsi rekursif:**
*  Fungsi rekursif harus memiliki base case, yang berfungsi untuk menghentikan pemanggilan terus menerus pada fungsi rekursif
*  Fungsi rekursif harus memiliki syntax yang dapar merubah state dari permasalahan, sehingga semakin lama solusi permasalahan menuju base case yang sudah dibuat
*  Fungsi rekursif harus memanggil dirinya sendiri

**Contoh**

Bilangan faktorial


```python
def factorialRekursif(bilangan):
  if bilangan <=1:
    return(1)
  else:
    return(bilangan * factorialRekursif(bilangan-1))

data = factorialRekursif(3)
print(data)
```

**Permasalahan Tower of Hanoi**

kita disuruh untuk memindahkan piringan yang tersusun rapi dari tower asal ke tower tujuan dan disediakan tower bantuan sebagai perantara memindahkan piringan tersebut.

Syarat pemindahan piringan:
*  piringan harus dipindahkan satu persatu
*  piringan berukuran kecil harus berada diatas piringan yang berukuran besar

image.png

```python
def towerOfHanoi(n, asal, pembantu, tujuan):
  if n == 1:
    print("pindahkan piringan " + str(n) + " dari " + asal + " ke " + tujuan)
  else:
    towerOfHanoi(n-1,asal, tujuan, pembantu)
    print("pindahkan piringan " + str(n) + " dari " + asal + " ke " + tujuan)
    towerOfHanoi(n-1, pembantu, asal, tujuan)

towerOfHanoi(3, "A", "B", "C")
```
