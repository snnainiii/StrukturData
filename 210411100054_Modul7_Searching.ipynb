### **Implementasi 2**

```python
# menginport function randint (memunculkan angka acak dengan tipe data interger)
from random import randint

# mengenerate n bil bulat 1 s/d n
def Data(n):
  list = []
  for i in range(500):
      data = randint(1,n)
      list.append(data)
  return list

# mengurutkan data menggunakan insertionsort
def insertionSort(listData):
  for outIter in range(1, len(listData)):
      key = listData[outIter]
      ind = outIter
      while (ind > 0 and listData[ind-1] > key):
          listData[ind] = listData[ind-1]
          ind = ind-1
          listData[ind] = key
  return listData

data = Data(101)
print("Data = ", data, "\n")
```

 **(a) Unordered Sequential Search**

```python
def unorderedSeqSearch(li, src):
  i = 0
  count = []
  while i < len(li):
    if li[i] == src:
      count.append(i)
    i += 1
  if count == []:
    return "{} is not Found".format(src)
  else:
    return "{} is in : {}".format(src,count)

print("Data = ", data, "\n")
print(unorderedSeqSearch(data,100))
print(unorderedSeqSearch(data,20))
print(unorderedSeqSearch(data,500))
```

**(b) Ordered Sequential Search**

```python
def orderedSeqSearch(listData, data):
  ind = 0
  search = []
  while ind < len(listData) and listData[ind] <= data:
    if listData[ind] == data:
      search.append(ind)
    ind += 1
  if len(search) == 0:
    return "{} is not Found, numberOfIteration : {}".format(data,ind)
  else :
    return "{} is in {}, numberOfIteration : {}".format(data,search,ind)

insertionSort(data)
print("Data = ", data, "\n")
print(orderedSeqSearch(data, 1))
print(orderedSeqSearch(data, 25))
print(orderedSeqSearch(data, 100))
print(orderedSeqSearch(data, 20))
print(orderedSeqSearch(data, 500))
print(orderedSeqSearch(data, 1000))
```

**(c) Binary Search**

```python
def binarySearch(listData,data):
  first = 0
  last = len(listData) - 1
  found = False
  ind = 0
  while first <= last and not found:
    midpoint = (first + last) // 2
    if listData[midpoint] == data:
      found = True
    elif listData[midpoint] > data:
      last = midpoint - 1
    else:
      first = midpoint + 1
    ind += 1

  if found:
    return "{} is in : {} , numberOfIteration : {}".format(data, midpoint, ind)
  else:
    return "{} is not Found, numberOfIteration : {}".format(data, ind)


data = Data(3000)
insertionSort(data)
print("Data = ", data, "\n")
print(binarySearch(data, 1))
print(binarySearch(data, 31))
print(binarySearch(data, 254))
print(binarySearch(data, 1320))
```
