### **Konsep_1**

**Perbedaan antara list, tupple dan dictionaries :**

**List**

*   list terdiri dari beberapa element atau anggota list dengan berbagai tipe data

*   elemen pada list dibuka dan ditutup dengan tanda kurung siku '[ ]'

**Tupples**

*   tipe data yang digunakan untuk menyimpan lebih dari satu nilai dalam satu variabel sekaligus.

*   dictionaries memiliki key dan value
*   value dapat berupa tipe data apapun, namun key harus berupa tipe data yang tidak dapat berubah

*   setiap item dipisahkan dengan tanda koma (,)
*   element pada dictionaries dibuka dan ditutup dengan kurung kurawal ({ })

*   untuk mengakses elemen dictionary, dapat menggunakan tanda kurung siku yang sudah dikenal bersama dengan key untuk mendapatkan nilainya

### **Konsep_2**
**Jenis looping dan perbedaanya, syntax umum dari looping tesebut**

Terdapat 2 looping yaitu : for loop dan while loop

**for loop**

for digunakan untuk mengulangi kode yang sudah diketahui banyak perulangannya, sehingga for memiliki kemampuan untuk mengulangi item dari urutan apapun

**while loop**

while  digunakan untuk perulangan yang memilliki syarat dan tidak tentu sebarapa banyak perulangannya, sehingga statement akan dieksekusi berkali-kali selama kondisi bernilai True

**sintax umum**









```python
# sintax umum for loop

for(inisialisasi, kondisi, perubahan):
  statement
```

**keterangan:**

*   inisialisasi : pernyataan untuk
menyatakan keadaan awal dari variabel kontrol
*   kondisi : ekspresi relasi yang menyatakan kondisi untuk kelaur dari perulangan
*   perubahan : pengatur perubahan nilai variabel kontrol



```python
# contoh for loop

hari = ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu']
for i in hari:
  print('sekarang hari :',i)
```

```python
# sintax umum while loop
while(kondisi):
  statement
```

```python
# contoh while loop
data = 1
while data < 10:
  print('data sekarang : ',data)
  data += 1
```

### **Tugas Praktikum_1**

```python
def listNum():
  stop = True
  hasil = []
  while stop:
      angka = input('input a number : ')
      if angka != "stop":
          hasil.append(int(angka))
      else:
          stop = False
  return hasil

def sum(num):
  hasil = 0
  for i in range(len(num)):
      hasil += num[i]
  return hasil

def avg(num1):
  hasil = 0
  for i in range(len(num1)):
      hasil += num1[i]
  Avg = hasil/len(num1)
  return Avg

def odd(num2):
  oddNum = []
  for i in range(len(num2)):
      if num2[i] % 2 == 1:
          oddNum.append(num2[i])
  return oddNum

def even(num3):
  evenNum = []
  for i in range(len(num3)):
      if num3[i] % 2 == 0:
          evenNum.append(num3[i])
  return evenNum

data = listNum()
print("All number : ", data)

rata_rata = avg(data)
print("Average of all number : ", rata_rata)

penjumlahan = sum(data)
print("sum of all number : ", penjumlahan)

bil_genap = even(data)
print("Even Number : ", bil_genap)
rata_genap = (avg(bil_genap))
print("Average of even number : ", rata_genap)
jumlah_genap = (sum(bil_genap))
print("Sum of even number : ", jumlah_genap)

bil_ganjil = odd(data)
print("Odd Number : ", bil_ganjil)
rata_ganjil = (avg(bil_ganjil))
print("Average of odd number : ", rata_ganjil)
jumlah_ganjil = (sum(bil_ganjil))
print("Sum of odd number : ", jumlah_ganjil)
```

```python
def createList(size):
    data = []
    for i in range(size):
        angka = int(input(f"Angka ke - {i} = "))
        data.append(angka)
    return data


def pers1(data):
    y = 0
    n = len(data)
    for i in data:
        y += i
    y = y/n
    return y


def pers2(data):
    z = 0
    y = pers1(data)
    for i in data:
        temp = i-y
        temp = temp ** 2
        z += temp
    z /= len(data)-1
    return z


ukuran = int(input('Ukuran list = '))
data = createList(ukuran)
print("data = ",data)
persamaan1 = pers1(data)
print('y : ',persamaan1)
persamaan2 = pers2(data)
print('z : ',persamaan2)
```

adalah matrik dimensi tunggal menyimpan daftar variabel dari tipe data yang sama, dimungkinkan untuk mengakses setiap variabel menggunakan indeks.

**Matrik 2 Dimensi**

adalah matrik multi-dimensi menyimpan data dalam format yang terdiri dari baris dan kolom.



```python
# matrik 1D
def buatMatrix1D(sz):
    hasil1 = []
    for i in range(sz):
        hasil1.append(int(input(" ")))
    print(hasil1)
    return hasil1

# matrik 2D
def buatMatrix2D(baris, kolom):
    hasil2 = []
    for i in range(baris):
        bar = []
        for j in range(kolom):
            bar.append(int(input(" ")))
        hasil2.append(bar)
    print(hasil2)
    return hasil2

#  jumlah matrik 1D
def jumlahMatrix1D(mat1, mat2):
    jumlah1 = []
    cek = len(mat1) == len(mat2)
    if cek:
        for i in range(len(mat1)):
            jumlah1.append(mat1[i] + mat2[i])
        return jumlah1
    else:
        return "ukuran matrix tidak sama"

# jumlah matrik 2D
def jumlahMatrix2D(mat1, mat2):
    jumlah2 = []
    cek = len(mat1) == len(mat2)
    if cek:
        for i in range(len(mat1)):
            bar = []
            for j in range(len(mat2)):
                bar.append(mat1[i][j] + mat2[i][j])
            jumlah2.append(bar)
    return jumlah2

# perkalian 1D
def kaliMatrix1D(mat1, mat2):
    kali1 = []
    cek = len(mat1) == len(mat2)
    if cek:
        for i in range(len(mat1)):
            kali1.append(mat1[i]*mat2[i])
        return kali1
    else:
        return "ukuran matrix tidak sama"

# perkalian 2D
def kaliMatrik2D(mat1, mat2):
    kali2 = []
    cek = len(mat1) == len(mat2)
    if cek:
        for i in range(len(mat1)):
            bar = []
            for j in range(len(mat2)):
                bar.append(mat1[i][j] * mat2[i][j])
            kali2.append(bar)
    return kali2


print("Pilih Jenis Matrik : ")
print("1. Matrik 1 Dimensi")
print("2. Matrik 2 Dimensi")

pilih = int(input("Masukkan pilihan : "))

if pilih == 1:
    size1 = int(input("Masukkan ukuran matrix 1 = "))
    mat1 = buatMatrix1D(size1)
    size2 = int(input("Masukkan ukuran matrix 2 = "))
    mat2 = buatMatrix1D(size2)

    hasilJml = jumlahMatrix1D(mat1, mat2)
    hasilKali = kaliMatrix1D(mat1, mat2)
    print("Hasil penjumlahan matrix = ", hasilJml)
    print("Hasil perkalian matrix = ", hasilKali)

elif pilih == 2:
    baris_a = int(input("Masukkan jumlah baris matrix A : "))
    kolom_a = int(input("Masukkan jumlah kolom matrix A : "))
    matrix_a = buatMatrix2D(baris_a, kolom_a)

    baris_b = int(input("Masukkan jumlah baris matrix B : "))
    kolom_b = int(input("Masukkan jumlah kolom matrix B : "))
    matrix_b = buatMatrix2D(baris_b, kolom_b)

    jmlMat = jumlahMatrix2D(matrix_a, matrix_b)
    kaliMat = kaliMatrik2D(matrix_a, matrix_b)
    print("Hasil penjumlahan matrix = ", jmlMat)
    print("Hasil perkalian matrix = ", kaliMat)
```
