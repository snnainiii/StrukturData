```python
# inisialisasi stack baru
def newStack():
 s=[]
 return (s)

# penambahan data
def push(s,data):
 s.append(data)

# pengurangan data
def pop(s):
 data=s.pop()
 return(data)

# untuk mengetahui data yang terletak diposisi top
def peek(s):
 return(s[len(s)-1])

# untuk mngetahui apakah stack dalam keadaan kosong
def isEmpty(s):
 return (s == [])

# untuk mngetahui panjang stack
def size(s):
 return(len(s))

def duplikat(huruf) :
  kata = newStack()

  for x in huruf :
    if kata == []:
      push(kata, x)
    elif x != kata[-1] :
      push(kata, x)

  hasil = "".join(kata)
  return hasil

duplikat("stackgaessssss")
```
